```python
from transformers import AutoModelForCausalLM, AutoTokenizer, TrainingArguments, Trainer
from peft import get_peft_model, LoraConfig, TaskType
from datasets import load_dataset

model_name = "microsoft/phi-2"

# Load model and tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
tokenizer.pad_token = tokenizer.eos_token

model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto", trust_remote_code=True)

# PEFT config
peft_config = LoraConfig(
    r=8,
    lora_alpha=32,
    task_type=TaskType.CAUSAL_LM,
    lora_dropout=0.1,
    bias="none"
)

model = get_peft_model(model, peft_config)
```

```python
def tokenize(example):
    # Format dataset into prompt template
    prompt_template = f'''
    ### Instructions
    {example["question"]}
    {example["question_body"]}

    ### Assistant
    {example["answer"]}
    '''
    input_ids = tokenizer(prompt_template, truncation=True, padding="max_length", max_length=512)
    input_ids["labels"] = input_ids["input_ids"].copy()
    return input_ids
```

```python
def tokenize_batch(batch):
    # Create prompt templates in batch
    prompts = [
        f"""### Instructions
{q}
{qb}

### Assistant
{a}"""
        for q, qb, a in zip(batch["question"], batch["question_body"], batch["answer"])
    ]

    # Tokenize all prompts at once
    tokenized = tokenizer(
        prompts,
        truncation=True,
        padding="max_length",
        max_length=512
    )

    # Copy input_ids to labels
    tokenized["labels"] = tokenized["input_ids"].copy()

    return tokenized
```

```python
# Training args
training_args = TrainingArguments(
    output_dir="/content/drive/MyDrive/Phi 2 finetune/phi2-finetuned",
    per_device_train_batch_size=4,
    num_train_epochs=3,
    logging_dir="./logs",
    save_strategy="steps",
    save_steps=500,
    fp16=True,
    logging_steps=50,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset
)

trainer.train()
```
